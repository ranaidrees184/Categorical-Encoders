```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv('cars.csv')
df.sample(5)
```

```python
x_train,x_test, y_train,y_test=train_test_split(df.iloc[:,:4],df.iloc[:,-1],test_size=0.3)
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

```python
ohe=OneHotEncoder(drop='first',sparse_output=False)
xtrainbrand_cost=x_train[['brand','km_driven']].values
```

```python
xohenew=ohe.fit_transform(x_train[['fuel','owner']])
yohenew=ohe.fit_transform(x_test[['fuel','owner']])
```

```python
print(xohenew.shape)
print(yohenew.shape)
```

```python
np.hstack((xtrainbrand_cost,xohenew)).shape
```

```python
np.hstack((x_test[['brand','km_driven']].values,yohenew)).shape
```
