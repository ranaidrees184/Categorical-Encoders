```python
import pandas as pd
```

```python
df=pd.read_csv('customer.csv')
```

```python
df.head()
df=df.iloc[:,2:]
df.sample(4)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test=train_test_split(df.iloc[:,:2],df.iloc[:,-1],test_size=0.3)
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

```python
from sklearn.preprocessing import OrdinalEncoder
OE=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])

x_train_en=OE.fit_transform(x_train)
y_test_en=OE.fit_transform(x_test)
```

```python
x_train_en=pd.DataFrame(x_train_en, columns=x_train.columns)
y_test_en=pd.DataFrame(y_test_en, columns=x_test.columns)
```
